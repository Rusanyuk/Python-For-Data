```python
import pandas as pd
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv"
housing_in_london = pd.read_csv(url)

  # Display information
def information():
  return housing_in_london.info()
information()

  # Show the number and list of unique zones in the Area column
def unique_areas():
  unique_areas = housing_in_london['area'].unique()
  unique_areas_number = len(unique_areas)
  areas_list = ""
  count = 0
  for area in unique_areas:
    count += 1
    areas_list += str(count)
    areas_list += ". "
    areas_list += area
    areas_list += "\n"
  unique_areas_title = f"\nThis dataframe contains {unique_areas_number} unique areas:\n{areas_list}"
  return unique_areas_title
test_1 = unique_areas()
print(test_1)

  # Display the Area with the smallest population
def min_population_size():
  min_population_size = housing_in_london['population_size'].min()
  return f"The minimum population is: \n{min_population_size}"
print(min_population_size())

  # Display the median salary for the entire period
def median_salary():
  median_salary = housing_in_london["median_salary"].median()
  return f"\nMedian salary is: \n{median_salary}"
print(median_salary())

# Display all dataframe columns
def display_columns():
  columns = housing_in_london.columns
  columns_list = ""
  count = 0
  for column in columns:
    count += 1
    columns_list += str(count)
    columns_list += ". "
    columns_list += column
    columns_list += "\n"
  return f"\nThis dataframe contains {len(columns)} columns: \n{columns_list}"
print(display_columns())

# Display main statistics
def main_statistics():
  statistics = housing_in_london.describe()
  return statistics
print(main_statistics())
```
